```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('red_filter_image.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Split channels
r, g, b = cv2.split(img)

# Estimate missing channels
# Normalize red and use it to boost green and blue
g_est = np.clip(r * 0.6, 0, 255).astype(np.uint8)
b_est = np.clip(r * 0.4, 0, 255).astype(np.uint8)

# Merge estimated channels
corrected = cv2.merge([r, g_est, b_est])

# Show original vs corrected
plt.subplot(1,2,1)
plt.title("Red Filtered Image")
plt.imshow(img)

plt.subplot(1,2,2)
plt.title("Corrected Image")
plt.imshow(corrected)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load and convert to grayscale from red channel
img = cv2.imread('red_filter_image.jpeg')
red_channel = img[:,:,2]

# Apply CLAHE (adaptive histogram equalization)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
enhanced = clahe.apply(red_channel)

plt.imshow(enhanced, cmap='gray')
plt.title("Enhanced Body Features")
plt.axis('off')
plt.show()
```
